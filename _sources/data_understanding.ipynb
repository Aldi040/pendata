# 🌟 Data Understanding

**Data Understanding** adalah tahap awal dalam proses *Data Mining* (Penambangan Data) yang bertujuan untuk mendapatkan wawasan awal mengenai data yang akan dianalisis. Tahap ini sangat penting karena membantu kita mengetahui kualitas data, menemukan pola awal, dan mengidentifikasi potensi masalah seperti data yang hilang (*missing data*) atau nilai ekstrem (*outliers*).

### 🛠️ Beberapa kegiatan dalam memahami data di antaranya adalah:

- 📊 **Pengumpulan Data**  
  Mengumpulkan data dari berbagai sumber seperti database MySQL, PostgreSQL, file CSV, API, dll.

- 🔍 ***Profiling* Data**  
  Menganalisis struktur data seperti tipe data, jumlah baris, kolom, dan ringkasan statistik (mean, median, modus, dll.).

- 📈 **Eksplorasi Data**  
  Menggunakan visualisasi seperti grafik, histogram, dan diagram pencar untuk mengamati pola, distribusi, dan hubungan antarvariabel.

- 🚨 **Identifikasi Masalah Data**  
  Menemukan *missing values*, duplikasi data, dan *outliers* yang bisa memengaruhi hasil analisis.

- 📚 **Pembersihan Data**  
  Menghapus atau mengisi *missing values*, menangani duplikasi data, dan memperbaiki format data agar siap untuk dianalisis.

````python
import pymysql

# Koneksi ke database
connection = pymysql.connect(
    host="mysql-34425cbd-irismysqlaldi.h.aivencloud.com",
    port=22476,
    user="avnadmin",
    password="AVNS_QsQ7Yf7zzcmrk83yFgg",
    db="irismysqlaldi",
    charset="utf8mb4",
    cursorclass=pymysql.cursors.DictCursor,
    connect_timeout=10,
    read_timeout=10,
    write_timeout=10
)

# Fungsi untuk membuat garis tabel
def print_mysql_table(results):
    if not results:
        print("Empty set (0 rows)")
        return

    # Ambil header kolom
    headers = results[0].keys()
    lengths = [len(str(header)) for header in headers]

    # Cek panjang maksimum setiap kolom
    for row in results:
        for i, value in enumerate(row.values()):
            lengths[i] = max(lengths[i], len(str(value)))

    # Garis horizontal
    def print_separator():
        print("+" + "+".join("-" * (length + 2) for length in lengths) + "+")

    # Cetak header
    print_separator()
    print("| " + " | ".join(f"{header.ljust(length)}" for header, length in zip(headers, lengths)) + " |")
    print_separator()

    # Cetak isi baris
    for row in results:
        print("| " + " | ".join(f"{str(value).ljust(length)}" for value, length in zip(row.values(), lengths)) + " |")

    print_separator()
    print(f"{len(results)} rows in set")

try:
    with connection.cursor() as cursor:
        cursor.execute("SELECT * FROM flowers")
        results = cursor.fetchall()
        print_mysql_table(results)
finally:
    connection.close()
````

## 📊 Profiling Data

**Profiling data** adalah proses untuk menganalisis dan meringkas karakteristik dasar data. Pada analisis ini, statistik deskriptif digunakan untuk memahami dataset secara menyeluruh. Beberapa metrik utama yang biasanya digunakan meliputi:

- 📈 **Mean (Rata-rata)**: Menghitung rata-rata dari setiap kolom numerik untuk mengetahui nilai tengah data.
- 🔽 **Min (Nilai Minimum)**: Mencari nilai terkecil untuk setiap kolom numerik, membantu mengenali titik data terendah.
- 🔼 **Max (Nilai Maksimum)**: Mencatat nilai terbesar dalam kolom numerik, penting untuk mengetahui titik data tertinggi.
- 🔢 **Count (Jumlah Data)**: Menghitung jumlah baris atau entri untuk setiap kategori tertentu, seperti menghitung jumlah sampel per kelas.

✨ **Profiling ini bertujuan untuk:**
1. 📊 **Memahami distribusi data**: Mengetahui bagaimana data tersebar dan apakah ada pola tertentu.
2. 🚨 **Mengidentifikasi outliers**: Nilai ekstrem bisa berdampak besar pada analisis, jadi penting untuk mendeteksinya.
3. ✅ **Memastikan kualitas data**: Menemukan data yang hilang, inkonsisten, atau duplikat sebelum melangkah ke analisis lebih dalam.

📋 Hasil dari proses profiling biasanya disajikan dalam tabel agar lebih mudah dianalisis dan dibandingkan.

## 📊 Tipe Data Understanding dalam Data Mining

Dalam proses data mining, tahap *Data Understanding* sangat penting untuk memahami karakteristik data yang kita miliki. Salah satu aspek utama adalah mengenali **tipe data** yang ada. Berikut adalah macam-macam tipe data yang biasa ditemukan dalam data mining:

### 🔹 1. **Data Numerik (Numerical Data)**
Data yang berupa angka dan bisa dioperasikan secara matematis. Tipe ini terbagi menjadi dua:
- **Diskrit (Discrete):** Nilai yang dapat dihitung, biasanya bilangan bulat, misalnya jumlah siswa dalam kelas.
- **Kontinu (Continuous):** Nilai yang tidak terbatas dalam rentang tertentu, misalnya suhu atau berat badan.

### 🔸 2. **Data Kategorikal (Categorical Data)**
Data yang merepresentasikan kategori atau label tertentu. Tipe ini juga terbagi menjadi dua:
- **Nominal:** Kategori tanpa urutan, misalnya jenis kelamin (Pria, Wanita).
- **Ordinal:** Kategori yang memiliki tingkatan atau urutan, misalnya tingkat pendidikan (SD, SMP, SMA).

### 🔹 3. **Data Biner (Binary Data)**
Data yang hanya memiliki dua kemungkinan nilai, sering kali digunakan untuk representasi *True/False* atau *Yes/No*. Contohnya, apakah seseorang lulus ujian (Lulus/Tidak Lulus).

### 🔸 4. **Data Waktu (Time-Series Data)**
Data yang dikumpulkan berdasarkan urutan waktu tertentu. Contohnya, data penjualan harian, suhu mingguan, atau pergerakan harga saham.

### 🔹 5. **Data Teks (Text Data)**
Data dalam bentuk teks bebas yang biasanya digunakan dalam *text mining*. Contohnya ulasan pelanggan atau komentar media sosial.

### 🎯 Pentingnya Memahami Tipe Data
Memahami tipe data sangat penting karena:
- **Metode Analisis:** Algoritma tertentu hanya bisa digunakan untuk tipe data tertentu.
- **Pra-pemrosesan:** Menentukan cara penanganan data, misalnya normalisasi untuk data numerik.
- **Visualisasi:** Memilih grafik atau plot yang sesuai untuk setiap tipe data.
